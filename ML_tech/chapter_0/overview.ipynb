1、什么是机器学习：
它的定义随着理论的发展而发展，60年代的推理，70年代的知识工程，80年代到现在的机器学习，实际上，即使是在 机器学习的专业人士中，也不存在一个被广泛认可的定义 来准确定义机器学习是什么或不是什么，现在我将告诉你一些大咖们尝试定义的示例。（）第一个机器学习的定义来自于1952 年, 被誉为“机器学习之父”IBM 的亚瑟·塞缪尔Arthur Samuell，对就是这位大牛（手指动作）。
 他定义机器学习为，在进行特定编程的情况下，给予计算机学习能力的领域。塞缪尔的定义 是建立在他编写了一个西洋棋程序。 这程序神奇之处在于，编程者自己并不是个下棋高手。 但因为他太菜了，于是就通过编程， 让西洋棋程序自己跟自己下了上万盘棋。通过观察 哪种棋盘布局位置会赢，哪种布局会输， 久而久之，这西洋棋程序明白了什么是好的布局， 什么样是坏的布局。然后程序通过学习后， 玩西洋棋的水平超过了塞缪尔。这绝对是令人注目的成果。 上述定义是个有点不正式的，也比较古老。
另一个年代近一点的定义，由来自卡内基梅隆大学，机器学习教父Tom Mitchell提出的， Tom定义了这么个啰嗦的机器学习概念，他说， 一个程序被认为能从经验E中学习，解决任务 T，达到 性能度量值P，当且仅当，有了经验E后，经过P评判，程序在处理 T 时的性能有所提升。
我认为他提出的这个定义很拗口，在西洋棋那例子中，经验e 就是 程序上万次的自我练习的经验，而任务 t 就是下棋。性能度量值 p呢， 就是它在与一些新的对手比赛时，赢得比赛的概率。
我们提个问题给大家思考： 以骚扰电话识别为例，Tom Mitchell的机器学习的定义中，任务 T 是什么？性能度量值P是什么？机器学习的概念也决定了它的种类会很多

二、机器学习基本流程
1、机器学习基本流程
（）开发应用机器学习的算法程序的步骤主要有六部：
（1）、收集数据：如制作爬虫，从RSS反馈或者API中得到信息
（2）、输入数据： 将数据集中管理，便于应用。
（3）、数据探索与处理：判断的导入数据是否有异常值、丢失等，确保输入格式是符合要求的，补充规范数据，方便操作。如采用python中的list或者一些特定的格式。
（4）、训练模型（监督学习）：数据分为测试数据与训练数据，分别将格式化数据输入到训练模型算法中，抽取知识或者信息，得到的知识需要存储为计算机可以处理的格式。注：如果是非监督学习算法不需要训练算法
（5）、模型测试：将测试数据输入训练好的模型中，进行模型准确性和可靠性评估， 
（6）、评估：用来优化模型，并对模型部署与整合构建机器学习应用系统。
最后根据应用产生的反馈数据对机器学习应用进行优化。




2、算法选择
在完成这些过程中，核心的问题是算法选择与实现问题，算法没有划坏之分，只有好用不好用之分，如何选择合适算法，是很多学生比较关切的问题？
我们可以简单地这样理解，预测目标变量的值---使用监督学习算法，其中目标变量类型：离散型（A\B\C，红\黑\黄）---分类，比如一组照片中分辨出哪个是猫哪个是狗问题；连续型数值型（0.0~100.00）---回归算法，比如说明天股票会涨还是会跌问题，台风未来发展趋势情况问题；不预测目标变量的值---使用非监督学习，将数据划分为离散的组---聚类，估计数据与每个组的相似程度---密度估计，比如抑郁症、剁手族的界定等问题，如果是唯一的需求，使用聚类算法，如果还需要估计数据与每个分组的相似度，还需要使用密度估计算法。


3.关键术语
为了便于更好的理解相关知识，这里我们学习一些必要专业术语。 首先是
（1）、训练集概念：训练模型首先要有数据，其中每条数据时关于一个事件或一个对象的描述，它们的集合即称训练集：他是用于训练机器学习算法的数据样本集合
（2）、训练数据和测试数据：使用训练样本集作为算法的输入，训练完成之后输入测试样本，一般采用交叉验证方式。
（3）、属性/特征：反映事件或对象在某方面的表现或性质的事项，描述的关键点称其为特征或属性。如鸟的体重、翼展、脚蹼、后背颜色、种属等。
（4）、标称型：标称型目标变量的结果只在有限目标集中取值，如真与假(标称型目标变量主要用于分类)
（5）、数值型：数值型目标变量则可以从无限的数值集合中取值，如0.100，42.001等 (数值型目标变量主要用于回归分析)
（6）、学习 /训练：从数据中学得模型的过程
（7）、目标变量：机器学习算法的预测结果
（8） 准确率：是我们最常见的评价指标，而且很容易理解，就是被分对的样本数除以所有的样本数，通常来说，正确率越高，分类器越好。
（9）查准率：表示被分为正例的示例中实际为正例的比例。
（10）查全率：是更为适用的性能度量。召回率是覆盖面的度量
（11）激活函数：一般是非线性函数，其作用是能够给神经网络加入一些非线性因素，使得神经网络可以更好地解决较为复杂的问题。
（12）交叉验证：是在机器学习建立模型和验证模型参数时常用的方法。就是重复的使用数据，用训练集来训练模型，测试集来评估模型的好坏。在此基础上，某次训练集中的样本，在下次可能成为测试集中的样本，也就是所谓的交叉。
（13）、知识表示：某些算法可以产生很容易理解的知识表示，而有些算法的知识可能只能被计算机所理解。
（14）模型：泛指从数据中学得的结果。有文献用"模型"指全局性结 果（例如一棵决策树），而用"模式"指局部性结果（例如一条规则）。
（15）泛化能力：学得模型适用于新样本的能力。
